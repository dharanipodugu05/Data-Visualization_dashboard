```python
1. Line Chart: Daily Sales Trends
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data if not already loaded
df = pd.read_csv("archive/time series data.csv")

# Add date column if it doesn't exist
if 'date' not in df.columns:
    df['date'] = pd.date_range(start='2023-01-01', periods=len(df), freq='D')

# Compute total sales
df['total_sales'] = df[['ProductP1', 'ProductP2', 'ProductP3', 'ProductP4', 'ProductP5']].sum(axis=1)

# Plot without grid lines
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['total_sales'], marker='o', linewidth=1.5)
plt.title("Total Product Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
# plt.grid(True)  # Grid removed
plt.tight_layout()
plt.show()
```

```python
 2. Temperature vs Total Sales (Scatter Plot)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("archive/time series data.csv")

# Create total sales column
df['total_sales'] = df[['ProductP1', 'ProductP2', 'ProductP3', 'ProductP4', 'ProductP5']].sum(axis=1)

# Plot without grid lines
plt.figure(figsize=(8, 5))
plt.scatter(df['temperature'], df['total_sales'], alpha=0.9, c='green')
plt.title("Temperature vs Total Sales")
plt.xlabel("Temperature")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
3. Bar Chart: Average Sales per Product

```

```python
product_means = df[['ProductP1', 'ProductP2', 'ProductP3', 'ProductP4', 'ProductP5']].mean()

product_means.plot(kind='bar', color='red', figsize=(8,5))
plt.title("Average Sales per Product")
plt.ylabel("Average Units Sold")
plt.xlabel("Product")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
4. Boxplot: Sales Distribution by Product
```

```python
df[['ProductP1', 'ProductP2', 'ProductP3', 'ProductP4', 'ProductP5']].plot(kind='box', figsize=(10,6))
plt.title("Sales Distribution by Product")
plt.ylabel("Units Sold")
plt.tight_layout()
plt.show()
```
